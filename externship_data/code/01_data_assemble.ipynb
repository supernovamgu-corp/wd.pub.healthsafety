```python
from glob import glob
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
date_tag = "202405"
```

```python
regions = glob("../../data/regional/*")
print(regions)
```

```python
data_1 = []
data_2 = []
data_3 = []

for r in regions:
    current_region = r.split("/")[-1]
    #1
    temp = pd.read_csv(r+"/{}/1_level_table.csv".format(date_tag))
    temp['source'] = current_region
    data_1.append(temp)
    #2
    temp = pd.read_csv(r+"/{}/2_level_table.csv".format(date_tag))
    # temp['source'] = current_region
    data_2.append(temp)
    #3
    temp = pd.read_csv(r+"/{}/3_level_table.csv".format(date_tag))
    # temp['source'] = current_region
    data_3.append(temp)
    
```

```python
df_level_1 = pd.concat(data_1)
df_level_2 = pd.concat(data_2)
df_level_3 = pd.concat(data_3)

print( df_level_1.shape, df_level_2.shape, df_level_3.shape, )
```

```python
print(df_level_3.shape, df_level_2.shape)
temp = df_level_2.drop_duplicates("url_inspection_full")
print(temp.shape, df_level_2.shape)
temp = temp[['url_inspection_full','url_entity_full']]
df_level_merged = pd.merge(df_level_3, temp, how='left', on ='url_inspection_full')

print(df_level_merged.shape)
```

```python
print(df_level_1.shape, df_level_merged.shape)
df_level_merged = pd.merge(df_level_merged, df_level_1, how='left', on ='url_entity_full')
print(df_level_merged.shape)
df_level_merged = df_level_merged.drop_duplicates()
print(df_level_merged.shape)
```
